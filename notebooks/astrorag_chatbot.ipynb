# 🌌 AstroRAG Interactive Chatbot

Evidence-aware retrieval over 408,590 arXiv astrophysics papers.

**Instructions:**
1. Run cell 1 to load the corpus (takes ~10s with cache, ~5 minutes fresh)
2. Run cell 2 to launch the chatbot UI
3. Use the tabs:
   - **🔬 Query** — Enter a research question and run the pipeline
   - **📊 Results** — View the structured summary with equations and numerical results
   - **💬 Chat** — Ask follow-up questions about the selected paper
   - **📚 Library** — Browse and reload previously processed papers

```python
# Cell 1 — Load corpus and initialise pipeline
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

from astrorag.data import DataLoader
from astrorag.data.models import LoadConfig
from astrorag.logger import setup_logging

setup_logging(level='INFO')

print('Loading corpus...')
config = LoadConfig(sample_size=408_590, use_cache=True, show_progress=False)
corpus = DataLoader(config=config).load()
print(f'✓ Loaded {corpus.n_papers():,} papers')
```

```python
# Cell 2 — Launch the chatbot UI
from astrorag.chat.widget import AstroRAGChatbot

bot = AstroRAGChatbot(corpus=corpus)
bot.show()
```

## Programmatic Access to Library

You can also access the context library directly without the UI:

```python
from astrorag.chat import get_library

lib = get_library()
print(f'Library has {len(lib.list_all())} papers')
print(f'Statistics: {lib.stats()}')

# List titles
for e in lib.list_all()[:5]:
    print(f'  {e.arxiv_id}: {e.title[:70]}')
```
